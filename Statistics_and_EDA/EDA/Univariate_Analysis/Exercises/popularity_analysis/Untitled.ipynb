```python
import sys, os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_media = pd.read_csv("popularity.csv")

df_media.head()

df_media.columns =df_media.columns.to_series().apply(lambda x: x.strip())
df_media.reindex()
```

```python
df_media.describe()
```

```python
#Find values below 78th percentile
df_media['shares'].quantile(0.78)
```

```python
df_media['shares'].quantile(np.linspace(.1, 1, 9,0))
```

```python
#Remove the outliers by removing the values which are more than 2 std away
mean_shares =df_media["shares"].mean()
std_shares = df_media["shares"].std()

print(mean_shares, std_shares)

df_shares = [x for x in df_media["shares"] if (x < (mean_shares+2*std_shares))]
df_shares = [x for x in df_shares if (x>(mean_shares-2*std_shares))]

pd.DataFrame(df_shares).mean()
```

```python
P = np.percentile(df_media["shares"], [0, 95])
print(P[0], P[1])
df_shares = df_media[(df_media["shares"] > P[0]) & (df_media["shares"] < P[1])]

df_shares["shares"].describe()
```
