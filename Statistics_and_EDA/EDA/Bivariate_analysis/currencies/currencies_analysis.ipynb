```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("currencies.csv")

df.head()
```

```python
corr = df.corr()

corr
```

```python
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```

```python
corr.style.background_gradient(cmap='coolwarm')
```
