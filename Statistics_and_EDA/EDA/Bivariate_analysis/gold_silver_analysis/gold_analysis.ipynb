```python
df_gold = pd.read_csv("EDA_Gold_Silver_prices.csv")

df_gold.head()

df_gold.info()
```

```python
def converter(x):
    try:
        return pd.datetime.strptime(x, '%b-%y')
    except ValueError:
        year, month = x.split('-')  # split by delimiter
        x = year.zfill(2) + '-' + month  # %y requires 0-padding
        return pd.datetime.strptime(x, '%y-%b')

res = df_gold["Month"].apply(converter)


df_gold
```
